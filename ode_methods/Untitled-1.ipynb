# Exercise 1: Wave Equation in flat space

We start considering the given wave equation in 1 + 1 dimensions, and natural units:
$$
\square \phi = - \frac{\partial^2 \phi}{\partial t^2} + \frac{\partial ^2 \phi}{\partial x^2}, \tag     1
$$
where obviously:
$$
\phi = \phi(t,x).
$$

We also introduce the given variables to reduce Equation $(1)$ (a second order hyperbolic Partial Derivatives Equation, hereafter PDE) into a system of first order PDEs:
$$
f \equiv \frac{\partial \phi}{\partial t} \text{ and } g \equiv \frac{\partial \phi}{\partial x}.
$$

1) **Part $(a)$**: 

Following an analogous procedure to the one used during lecture, we can reduce  Equation $(1)$ into the following system:
\begin{cases}
\frac{\partial f}{\partial t} = \frac{\partial g}{\partial x} \\
\frac{\partial g}{\partial t} = \frac{\partial f}{\partial x}
\end{cases}
to, then, be solved with the inclusion of an extra equation to recover $\phi$, e.g. $\frac{\partial \phi}{\partial t} = f$.

We now rewrite the system of the first two _coupled_ equations in a matrix form:
$$
\frac{\partial}{\partial t} \begin{pmatrix}f \\ g \end{pmatrix} = \begin{pmatrix} 0 & 1 \\ 1 & 0\end{pmatrix}
\frac{\partial}{\partial x}\begin{pmatrix}f \\ g \end{pmatrix},
$$
where we call for convinience:
$$
U \equiv \begin{pmatrix}f \\ g \end{pmatrix} \text{ and } A \equiv \begin{pmatrix} 0 & 1 \\ 1 & 0\end{pmatrix},
$$
obtaining the much nicer expression:
$$
\frac{\partial}{\partial t} U = A \frac{\partial}{\partial x} U. \tag{2}
$$

Recalling the long lost linear algebra notions, one can find the eigenvalues and eigenvectors of this system starting from saying that we want to satisfy the expression:
$$
Av_\lambda = \lambda v_\lambda, \text{ i.e.:}\\
\left(A-\lambda I \right) v_\lambda = 0
$$
where $v_\lambda$ is an eigenvector ($v_\lambda\neq 0$), $\lambda$ an eigenvalue, and $I$ is te identity matrix (I sadly didn't find the symbol I prefer of the 1 with two bars, sigh).
Since we want to obtain zero even if the vector is non zero, we need to ask that the matrix in parenthesis is not invertable, i.e.:
$$
\text{det}\left(A-\lambda I \right)=0.
$$
This implies: 
$$
\begin{vmatrix}
-\lambda & 1 \\
1 & \lambda
\end{vmatrix} = 0,
$$
which leads to the following characteristic polynomial:
$$
\lambda^2 - 1 = 0,
$$
with the two solutions $\lambda_+ = 1$ and $\lambda_- = -1$.

We can now find the eigenvectors associated for $\lambda_\pm = \pm 1$, by solving the system:
$$
\begin{pmatrix}
\mp 1 & 1 \\
1 & \mp 1
\end{pmatrix} \begin{pmatrix} v_1 \\ v_2 \end{pmatrix} = 0,
$$
which leads to:
\begin{cases}
v_2 = \pm v_1 \\
v_1 = \pm v_2
\end{cases}
and, thus, to the two eigenvectors:
$$
v_+ = \begin{pmatrix} 1 \\ 1 \end{pmatrix} \text{ and } v_- = \begin{pmatrix} 1 \\ -1 \end{pmatrix},
$$
associate respectively to $\lambda_+ = 1$ and $\lambda_- = -1$.
In order to decouple the system, and identify the ingoing and outgoing modes as requested, we can now introduce the so-called _characteristic variables_:
$$
\begin{cases}
u_+ = v_+ \cdot U = v^T_{+i} U_i = f + g \\
u_- = v_- \cdot U = v^T_{-i} U_i = f - g
\end{cases}.
$$
With some algebra, we can rewrite the system in Equation $(2)$ in terms of the characteristic variables, obtaining:
\begin{cases}
\frac{\partial u_+}{\partial t} - \frac{\partial u_+}{\partial x} = 0 \\
\frac{\partial u_-}{\partial t} + \frac{\partial u_-}{\partial x} = 0
\end{cases}.
This set of equations describes respectively: 
- a wave $u_+$ propagating at velocity $-1$, on $t+x=\text{const}$ lines, i.e. propagating to the **left**;
- a wave $u_-$ propagating at velocity $+1$, on $t-x=\text{const}$ lines, i.e. propagating to the **right**.

To now identify the _ingoing_ and _outgoing_ modes, we need to consider the boundary conditions of the problem. Assume we are interested in the solution of the PDE in the domain $x\in[x_0,x_1]$, then we can say that:
- the mode $u_+$ is **outgoing** at the boundary $x=x_0$ and **ingoing** at the boundary $x=x_1$;
- the mode $u_-$ is **ingoing** at the boundary $x=x_0$ and **outgoing** at the boundary $x=x_1$.
-------------------------------------------------------------------------------------------------------------------------------------

2) **Part $(b)$**: 

The first part of the question  refers to "Reflective Boundary Conditions" as equivalent to put outgoing modes vanishing at such boundary.
As such, we can easily see the implication of this condition on the two boundaries:
- at the boundary $x=x_0$, we have $u_+(t,x_0) = 0, \forall t$, which implies $$f(t,x_0) + g(t,x_0) = 0, \forall t$$ which for $\phi$ simply translates into$$\frac{\partial \phi}{\partial t}(t,x_0) + \frac{\partial \phi}{\partial x}(t,x_0) = 0, \forall t.$$
- at the boundary $x=x_1$, we have $u_-(t,x_1) = 0, \forall t$, which implies $$f(t,x_1) - g(t,x_1) = 0, \forall t$$ which for $\phi$ simply translates into$$\frac{\partial \phi}{\partial t}(t,x_1) - \frac{\partial \phi}{\partial x}(t,x_1) = 0, \forall t.$$

The second part refers to completely "Outgoing Boundary Conditions", which can be implemented by simply asking that the ingoing modes vanish at the boundaries (no waves from infinity), i.e.:
 - at the boundary $x=x_0$, we have $u_-(t,x_0) = 0, \forall t$, which implies $$f(t,x_0) - g(t,x_0) = 0, \forall t$$ which for $\phi$ simply translates into$$\frac{\partial \phi}{\partial t}(t,x_0) - \frac{\partial \phi}{\partial x}(t,x_0) = 0, \forall t.$$
 - at the boundary $x=x_1$, we have $u_+(t,x_1) = 0, \forall t$, which implies $$f(t,x_1) + g(t,x_1) = 0, \forall t$$ which for $\phi$ simply translates into$$\frac{\partial \phi}{\partial t}(t,x_1) + \frac{\partial \phi}{\partial x}(t,x_1) = 0, \forall t.$$

3) **Part $(c)$ and $(d)$**:
##### **Code for solving the PDE as a first order system (Part ($c$)) with given initial, and boundary conditions (Part ($d$))**.

We solve the system using the **Method of Lines**:

1. Replace spatial derivatives with finite differences.
2. Obtain a system of ODEs in time.
3. Integrate using a Runge–Kutta 4 (RK4) scheme.

##### **Initial Data**.

We evolve a **Gaussian pulse moving to the left**

$$
\phi(x,0) = \exp\!\left(-\frac{(x-x_c)^2}{2\sigma^2}\right),
$$

with initial conditions for $f$ and $g$ chosen so that the pulse propagates to the left, and $x_c\equiv\frac{x_1+x_0}{2}$ is the center of the pulse.

##### **Boundary Conditions**

We test three types of boundary conditions:

- **Periodic** s.t. $u_\pm(t,x_0) = u_\pm(t,x_1), \forall t$;
- **Totally reflective** (on both boundaries);
- **Mixed:** reflective on the left, outgoing on the right.

These are imposed using the preveasuly discussed characteristic variables $u_\pm = f \pm g$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dx_periodic(un, dx):
    """Centered derivative with periodic boundary conditions."""
    return (np.roll(un, -1) - np.roll(un, 1)) / (2 * dx)

def dx_upwind(un, dx, direction):
    """
    First-order upwind derivative.
    direction = +1  -> wave moving to the right
    direction = -1  -> wave moving to the left
    """
    if direction > 0:
        return (un - np.roll(un, 1)) / dx 
    else:
        return (np.roll(un, -1) - un) / dx
```

```python
def runge_kutta(tn, un, dx, h = 0.1, bc = 'periodic'):
    """
    Runge-Kutta 4th order method for solving ODEs.
    Parameters:
    h : float
        step size
    tn : float
        current time
    un : array_like
        current state vector

    Returns:
    un1 : array_like
        state vector at next time step
    """

    k1 = h*rhs(tn, un, dx, bc)
    k2 = h*rhs(tn + h/2, un + k1/2, dx, bc)
    k3 = h*rhs(tn + h/2, un + k2/2, dx, bc)
    k4 = h*rhs(tn + h, un + k3, dx, bc)

    un1 = un + 1/6*k1 + 1/3*k2 + 1/3*k3 + 1/6*k4
    return un1

# Converting the second-order PDE into a system of first-order PDEs
def rhs(tn, y, dx, bc):
    """Right-hand side of the first-order wave system."""
    
    # 1. Compute Base Derivatives (Centered / Periodic)
    dy_dx = dx_periodic(y, dx)
    
    # 2. Assign default equations (dt_phi = f, dt_f = dg_dx, dt_g = df_dx)

    # 3. Overwrite Boundaries for Non-Periodic cases
    if bc == "periodic":
        # Nothing to do! dx_periodic already handled the edges correctly.
        pass

    elif bc == "mixed":
        # --- LEFT (Reflective) ---
        # phi = 0 -> f = 0 -> dt_phi = 0, dt_f = 0
        y[0,0] = 0.
        y[1,0] = 0.
        y[2, 0] = dx_upwind(y[:, 1], dx, direction = -1)[0]

        # --- RIGHT (Sommerfeld / Outgoing) ---
        # Wave moves Right. Transport Eq: dt_f + dx_f = 0  =>  dt_f = - dx_f
        # We need dx_f using a BACKWARD difference.
        
        # Calculate the specific upwind derivatives needed at the edge
        df_backward = dx_upwind(f, dx, direction=+1)[-1]
        dg_backward = dx_upwind(g, dx, direction=+1)[-1]
        dy_backward = dx_upwind(y, dx, direction=+1)[-1]

        y[0, -1] = y[1, -1]
        y[1, -1] = -dy_backward[0]
        y[2, -1] = - dy_backward[1]

    elif bc == "reflective":
        # --- LEFT (Reflective) ---
        y[0,0] = 0.
        y[1,0] = 0.
        y[2, 0] = dx_upwind(y[:, 1], dx, direction=-1)[0] # Forward diff

        # --- RIGHT (Reflective) ---
        y[0,-1] = 0.
        y[1,-1] = 0.
        # dt_g = df/dx. We need a backward difference at x=L.
        y[2, 0] = dx_upwind(y[:, 1], dx, direction=-1)[-1] # Backward diff

    return y

    # Function to enforce reflective BCs
def apply_bc(y, bc):
    if bc == "periodic":
        # Periodic BCs are naturally handled in the rhs function, so we do nothing here
        pass
        
    elif bc == "mixed":
        # Left Reflective: Force hard zeros
        y[0,0] = 0
        y[1, 0]   = 0
        # Right Outgoing: Handled by rhs (no hard constraint)
        
    elif bc == "reflective":
        y[0,0], y[1,0]   = 0, 0
        y[0,-1], y[1, -1] = 0, 0
        
    return y
```

```python
a = 1.0
cf = 0.3
t_final = 20.0
L = 10.0     
nx = 400 
x = np.linspace(0., L, nx)
dx = x[1]-x[0]

dt = cf*dx
time = np.linspace(0., t_final, int(t_final/dt))
steps = int(t_final/dt)

bc = "periodic"
method = "upwind" 
```

```python
x0 = 0.5
sigma = 0.1
phi = np.exp(-(x-x0)**2/(2*sigma**2))
g = dx_periodic(phi, dx)

f = g.copy()

u = np.array([phi, f, g])
```

```python
t = 0.
u_list = []

for idx, t in enumerate(time):
    u = runge_kutta(t, un = u, dx = dx,h = dt, bc = 'periodic')
    t = t+dt

    if idx % 5 == 0.:
        u_list.append(u)
    
u_list = np.array(u_list)[::20]

phi_array = u_list[:, 0]
f_array = u_list[:, 1]
g_array = u_list[:, 2]
```

```python
# Plot
from IPython.display import clear_output
for frame in phi_array:
    clear_output(wait=True)
    plt.plot(x, frame)
    #plt.ylim(0., 15.1)
    plt.title(f"Wave evolution — BC = {bc}")
    plt.xlabel("x")
    plt.ylabel("phi")
    plt.show()
```

```python
# Comparison of different boundary conditions
for bc in ["periodic", "reflective", "mixed"]:
    phi= 10.*np.exp(-(x-x0)**2/(2*sigma**2))

    dx = x[1]-x[0]
    g = dx_periodic(phi, dx)

    f = g.copy()

    t = 0.
    while t < t_final:
            u = runge_kutta(t, un = [phi, f, g], h = dt, bc = 'periodic')
            t = t+dt

    plt.plot(x, u[0], label=bc)

plt.legend()
plt.title("Comparison of boundary conditions")
plt.show()
```
